# We have given two dataset, lets discuss 

#1.Play store data (first dataset)
This contains data on the Google Play applications. It has 10,841 rows of data witch has following columns:

<u>App Category:</u> Category of the app. This could be beauty, business, medical,art and design entertainment, education...etc.

<u>Rating:</u> It has How users rate the app out of 5, with 1 being the lowest rating and 5 being the highest.

<u>Reviews:</u> number of user reviews each app has received.

<u>Size:</u> The memory size needed to install the application.

<u>Installs:</u> The number of times each application has been installed by users.

<u>Type:</u> Whether the app is free or a paid app.

<u>Price:</u> The price of the app.

<u>Content Rating:</u> This column specifies the intended audience for the app. Can be for teens, mature audience, or everyone.

<u>Genres:</u> The sub-category for each app. Example: for the Education category, this could be Education: Pretend Play, for example.

<u>Last Updated:</u> Release date of the most recent update for the app.

<u>Current Ver:</u> The app's current version.

<u>Android Ver:</u> The oldest version of Android OS supported by the app.

#2.User reviews dataset (second dataset)
This file contains the result of the sentiment analysis conducted by the dataset creator. It has 64,295 rows of data with the following columns:

<u>App:</u> Name of the app.

<u>Translated_Review:</u> Either the original review in English, or a translated version if the orignal review is in another language.

<u>Sentiment:</u> The result of the sentiment analysis conducted on a review. The value is either Positive, Neutral or Negative.

<u>Sentiment_Polarity:</u> A value indicating the positivity or negativity of the sentiment, values range from -1 (most negative) to 1 (most positive). Sentiment polarity for an element defines the orientation of the expressed sentiment, i.e. it determines if the text expresses the positive, negative or neutral sentiment of the user about the entity in consideration.

<u>Sentiment_Subjectivity:</u> A value from 0 to 1 indicating the subjectivity of the review. Lower values indicate the review is based on factual information, and higher values indicate the review is based on personal or public opinions or judgements.

We can see that the values in the Size column contains data with different units. 'M' stands for MB and 'k' stands for KB. To analyse this column we will convert all the units to MB to easily understand and analyze.

```python
#define a function to convert kb to mb
def convert_kb_to_mb(val):
  '''
  This function converts all the valid entries in KB to MB and returns the result in float datatype.
  '''
  try:
    if 'M' in val:
      return float(val[:-1])
    elif 'k' in val:
      return round(float(val[:-1])/1024, 4) #as we know 1MB = 1024KB
    else:
      return val
  except:
    return val
```

```python
#apply above function to convert Size to single unit MB
ps_df['Size'] = ps_df['Size'].apply(lambda x: convert_kb_to_mb(x))
```

```python
#we need to deal with 'Varies with device' lest replace it with NaN
ps_df['Size'] = ps_df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
```

```python
# change dtype object to float
ps_df['Size'] = ps_df['Size'].apply(lambda x: float(x))
```

```python
#groupby rating based on category nd using mean to find rating average 

rating_avg_per_category= pd.DataFrame(ps_df.groupby(['Category'])['Rating'].mean().sort_values(ascending=False))

rating_avg_per_category
```

we can group the rating as follows:
1. 4-5: Top rated
2. 3-4: Above average rated
3. 2-3: Average rated 
4. 1-2: Bad rated
 Fine lets create a column as <code>Rating_Group</code> in main Dataframe and apply above filter


```python
#lets define a function to apply above filter
def rating_app(rating):
  ''''
  This function use to categories the rating from 1 to 5 as
  Top_Rated,Above_Average,Average & Bad_Rated.
  ''' 
  if rating>=4:
    return 'Top_Rated'
  elif rating >3 and rating <4:
    return 'Above_Average'
  elif rating >2 and rating <3:
    return 'Average'
  else:
    return 'Bad_Rated'
```

<b>Lets apply above function to the 'Rating' attribute and save the output column as <code>Rating_Group</code> in the ps_df dataframe.<b>

```python
#applying function to rating.
ps_df['Rating_Group']=ps_df['Rating'].apply(lambda x: rating_app(x))
```

```python
#lets plot the average app rating
ps_df['Rating_Group'].value_counts().plot.bar(figsize=(15,5), color =['g','b','r','r'])
plt.xlabel('Rating Group', fontsize= 15)
plt.ylabel('Number of apps', fontsize= 15)
plt.title('Average app rating', fontsize=20)
plt.legend()
```

from abaove graph we can conclude 
> 1. For all the category the average rating is above 4 stars.
> 2. Most of the application are "Top_Rated" as we see earler in the analysis.

# Q.4: Find top 10 apps in any category based on the their install count.

```python
#create function
def top_ten_apps(cat_name):

  '''This func takes category as argument and returns 
      top 10 apps based om install
  '''
  cat_name=cat_name.upper()
  top10app=ps_df[ps_df['Category']==cat_name].sort_values(by='Installs',ascending=False).head(10)
  #plot visual
  plt.figure(figsize=(15,6),dpi=100)
  plt.title('Top 10 install apps',size=20)
  graph=sns.barplot(x=top10app.App, y= top10app.Installs, palette ='colorblind')
  graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right')
```

```python
#lets apply above function on 'MEDICAL' category
top_ten_apps('MEDICAL')
```

we can now fetch up top 10 apss from any category based on the number of install using our customized function
